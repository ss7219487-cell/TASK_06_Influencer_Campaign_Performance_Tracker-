```python
df["Engagement_Rate"] = (
    (df["Likes"] + df["Comments"] + df["Shares"])
    / df["Reach"]
) * 100
```

```python
df["ROI"] = (
    (df["Revenue"] - df["Spend"])
    / df["Spend"]
) * 100
```

```python
def influencer_size(followers):
    if followers < 100000:
        return "Micro Influencer"
    elif followers < 1000000:
        return "Macro Influencer"
    else:
        return "Mega Influencer"

df["Influencer_Size"] = df["Followers"].apply(influencer_size)
```

```python
df[["Followers","Influencer_Size"]].head()
```

```python
print(df.groupby("Influencer_Size")[["Revenue","ROI","Engagement_Rate"]].mean())
```

```python
print(df.columns)
```

```python
df["ROI_Category"] = df["ROI"].apply(
    lambda x: "High ROI" if x >= 100 else "Low ROI"
)
```

```python
df[["ROI","ROI_Category"]].head()
```

```python
average_engagement = df["Engagement_Rate"].mean()

df["Engagement_Category"] = df["Engagement_Rate"].apply(
    lambda x: "High Engagement"
    if x >= average_engagement
    else "Low Engagement"
)
```

```python
df[["Engagement_Rate","Engagement_Category"]].head()
```

```python
df["Cost_Efficient"] = df.apply(
    lambda row: "Yes"
    if row["Revenue"] > row["Spend"]
    else "No",
    axis=1
)
```

```python
df[["Spend","Revenue","Cost_Efficient"]].head()
```
